```python
#Six columns with null values are replaced with appropriate values
import numpy as np
import pandas as pd
data=pd.read_csv("netflix.csv")
data["duration"].fillna(0, inplace=True)
data["cast"].fillna("Not mentioned", inplace=True)
data["country"].fillna("Not mentioned", inplace=True)
data["date_added"].fillna("Not Known", inplace=True)
data["rating"].fillna("Not Sure", inplace=True)
data["director"].fillna("Unknown", inplace=True)
data
```

```python
#Unpacking multiple values in column 'listed_in'
import numpy as np
import pandas as pd
data=pd.read_csv("netflix.csv")
data=data.assign(listed_in=data['listed_in'].str.split(',')).explode('listed_in').reset_index(drop=True).sort_values(by='duration')

data
```

```python
#Unpacking multiple values from rows of the column "cast"
casting=data.assign(cast=data['cast'].str.split(',')).explode('cast').reset_index(drop=True).sort_values(by='title')
sd=casting.sort_values('cast')
sd['cast'].unique()
```
